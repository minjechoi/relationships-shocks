```python
import os
from os.path import join
import pandas as pd
import ujson as json
import numpy as np
```

## Count of potential shock tweets

- Only corresponds to reg.3 tweets (reg.1 and reg.2 exist in the cavium server)

```python
load_dir = '/shared/2/projects/relationships-shocks/working-dir/filtered-tweets-cavium/'
out = []
for file in sorted(os.listdir(load_dir)):
    if file.startswith('reg3.') and (not 'dup' in file):
        cat = file.split('.')[1]
        with open(join(load_dir,file)) as f:
            for ln,line in enumerate(f):
                obj = json.loads(line)
                text = obj['text']
                date = obj['timestamp']
                out.append((cat,date,text))
df = pd.DataFrame(out,columns=['category','date','text'])
print(len(df))
df = df.dropna().drop_duplicates()
print(len(df))
df.groupby('category').count()
```
